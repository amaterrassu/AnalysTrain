```python
def custom_hist(df : pd.Series):
  
  fig = plt.figure(figsize = (7, 7))
  ax = fig.add_subplot()
  plt.hist(df, density = True, color = 'red', edgecolor = 'skyblue', bins = 20, label = 'hist')
  plt.title(f'Histofram of {df.name}')
  plt.xlabel('Ox')
  plt.ylabel('Oy')
  plt.xticks(rotation = 90)
  plt.minorticks_on()
  plt.grid(which='major', ls='--')
  plt.grid(which='minor', linewidth=1)
  plt.show();

def describe_columns(df : pd.Series):

  print(f'Количество пропусков {df.isna().sum()}')
  print(f'Тип : {df.dtypes}')

  
```
